```python
cdf = pd.read_csv('/home/shrey/ENCODE2/team5/Analysis/E-GEOD-47399-raw-counts.tsv', sep = '\t')
```

```python
cdf.drop(columns=['Gene ID'], axis = 1, inplace = True)
cdf.dropna(inplace = True)
mask = cdf['Gene Name'].value_counts() > 1
genestoremove = mask[mask == True].index
cdf = cdf[~cdf['Gene Name'].isin(genestoremove)]
cdf.reset_index(drop = True, inplace = True)
cdf.to_csv('/home/shrey/ENCODE2/team5/Analysis/E-GEOD-47399-raw-counts-somethingintheway.tsv', sep = '\t', index = False)
```
